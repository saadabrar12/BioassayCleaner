## Parsing the bioassay data
The target specific information is spread throughout the bioassay record. The fields that hold the information relevant to the targets are mined and parsed from the raw JSON bioassay data.  After reviewing couple of bioassays, it is seen that the fields that needs to be parsed are 
* Target field
    * mol_id fields (Contained information about the GIs)
    * Target comment fields (Usually contains the Uniprot ID)
* Xref field
    * Protein_gi (It holds the relevant GIs for the target and hence can be parsed to eke out the target identity)
    * gene (Some bioassays hold the Gene ID(Entrez) that is a direct reference to the target referred to in the bioassay)

The panel assays required special attention and were also parsed for information like the target fields and the xref fields.

The size of the overall dataset and the distributions of the sizes of the dataset is essential to get a hold of the time to parse the whole dataset.
```python
import os
def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            # skip if it is symbolic link
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)

    return total_size

#print(get_size(), 'bytes')

path = 'C:\\Users\\saad6\\Downloads\\CSC-830_Spring_2020\PubChem\\JSON\\Folders\\'
os.chdir(path)
folders = os.listdir()

sizes = []

for folder in range(len(folders)):
    sizes.append(get_size(folders[folder]))
```

```python
import os
import gzip
import json
import pickle
import time
class Bioassay:
    def __init__(self,json_data,num):
        self.aid = num
        self.aname = json_data['PC_AssaySubmit']['assay']['descr']['name']
        
        try:
            self.adesc = ''.join(json_data['PC_AssaySubmit']['assay']['descr']['description'])
        except KeyError as err:
            self.adesc = None
        
        #List of target IDs found in assay
        self.target_id_list=[]
        
        # Gene ID present in xrefs
        self.is_gene_xref = False
        
        # Gene ID
        self.gene_id_list = []
        
        # Target comments
        self.target_comment = []
        
        # Panel target genes
        self.panel_gene_id = []
        
        #Panel GI or target IDs
        self.panel_target_ID = []
        '''
        #if(len(json_data['PC_AssaySubmit']['assay']['descr']['description'])==3):
        #    self.adesc = json_data['PC_AssaySubmit']['assay']['descr']['description'][2]
        #else:
        #    self.adesc = json_data['PC_AssaySubmit']['assay']['descr']['description'][0]
        #try:
        #    self.tid_definitions = json_data['PC_AssaySubmit']['assay']['descr']['results']
        #except KeyError as err:
        #    self.tid_definitions = None
        #try:
        #    self.tid_data_table = json_data['PC_AssaySubmit']['assay']['descr']['data']
        #except KeyError as err:
        #    self.tid_data_table = json_data['PC_AssaySubmit']['data']
        '''
        
        # Target info [1st-in target field, 2nd in xref, 3rd in assay comment, 4th in target comment]
        self.target_info_bitmap = [False, False] 
        
        try:
            self.target = json_data['PC_AssaySubmit']['assay']['descr']['target']
            self.target_info_bitmap[0] = True
            self.parse_target()
            
        except KeyError as err:
            self.target = None
            self.target_comment = None
        
        #Xrefs
        try:
            self.xrefs = json_data['PC_AssaySubmit']['assay']['descr']['xref']
            self.parse_xrefs()
        except KeyError as err:
            self.xrefs = None
        # Comment fields
        try:
            self.comment = json_data['PC_AssaySubmit']['assay']['descr']['comment']
        except KeyError as err:
            self.comment = None
        #Handling Panel Assays
        try: 
            self.is_panel = bool(json_data['PC_AssaySubmit']['assay']['descr']['is_panel'])
            if(self.is_panel == True):
                self.panel_info = json_data['PC_AssaySubmit']['assay']['descr']['panel_info']
                self.parse_panel_info()
            else:
                self.panel_info = None
        except KeyError as err:
            self.is_panel = False
            self.panel_info = None
    
    #Parsing target
    def parse_target(self):
        #Adding the target_id to the target list
        for i in self.target:
            self.target_id_list.append(i['mol_id'])
            self.target_info_bitmap[0] = True
            try: 
                self.target_comment.append(i['comment'])
            except KeyError as err:
                continue
            #self.target_id_name.append(i[])
        #self.target_comment = " ".join(self.target_comment)
    # Parsing xrefs
    def parse_xrefs(self):
        #Parsing gene IDs and GI
        for i in self.xrefs:
            # Gene IDs
            try:
                self.gene_id_list.append(i['xref']['gene'])
                self.is_gene_xref = True
            except KeyError as err:
                self.gene_id = None
            # Protein GI
            try:
                protein_gi = i['xref']['protein_gi']
                if protein_gi not in self.target_id_list:
                    self.target_id_list.append(i['xref']['protein_gi'])
                self.target_info_bitmap[1] = True
            except KeyError as err: 
                continue
    
    #parsing panel info
    def parse_panel_info(self):
        for i in self.panel_info['member']:
            try:
                for j in i['target']:
                    if j['mol_id'] not in self.panel_target_ID:
                        self.panel_target_ID.append(j['mol_id'])
            except KeyError as err:
                pass
            try:
                for j in i['xref']:
                    try:
                        if j['gene'] not in self.panel_gene_id:
                            self.panel_gene_id.append(j['gene'])
                    except KeyError as err:
                        pass
            except KeyError as err:
                pass
        
    
```

## Crawling through the parsed data

The crawler module have found a total of **774** bioassays that are relevant to Human androgen receptors(hAR). The IDs of the relevant bioassays are listed. 

```python
list_of_AR_Bioassays = []

for b in assays:
    if len(b.target_id_list)!=0:
        for i in b.target_id_list:
            if i in AR_GI and b not in list_of_AR_Bioassays:
                list_of_AR_Bioassays.append(b)
    if b.is_gene_xref:
        if AR_Gene_Id in b.gene_id_list and b not in list_of_AR_Bioassays:
            list_of_AR_Bioassays.append(b)
    if b.is_panel:
        for i in b.panel_target_ID:
            if i in AR_GI and b not in list_of_AR_Bioassays:
                list_of_AR_Bioassays.append(b)
        
        if AR_Gene_Id in b.panel_gene_id and b not in list_of_AR_Bioassays:
                list_of_AR_Bioassays.append(b)
print(len(list_of_AR_Bioassays))


ids = [i.aid for i in list_of_AR_Bioassays]
ids
```

Finding info underlying in the assay comment section and the target comment sections

```python
import re

def getNumbers(str): 
    array = re.findall(r'[0-9]+', str) 
    return array

def UniProtID(str):
    array = re.findall(r'P[0-9]+', str)
    return array

bioassays_found_comment = []

for b in assays:
    if b.target_comment:
        string_val=""
        for a in b.target_comment:
            for a1 in a:
                string_val+=a1
        #string_val = " ".join(a1) for a in b.target_comment for a1 in a
        #print(string_val)
        arr = UniProtID(string_val)
        for a in arr:
            if a == AR_UniProt_Id:
                bioassays_found_comment.append(b.aid)
                break

print(len(bioassays_found_comment))
bioassays_found_comment
```
